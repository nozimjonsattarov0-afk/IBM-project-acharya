```python
import pandas as pd
df = pd.read_csv('../data/processed/cleaned_data.csv')
median_spend = df['Purchase_Amount'].median()
df['Is_High_Spender'] = (df['Purchase_Amount'] > median_spend).astype(int)
df = pd.get_dummies(df, columns=['Purchased_Product'], prefix='Prod')

print("New features created. Shape of data:", df.shape)
df.head()
```

```python
bins = [18, 25, 35, 50, 100]
labels = ['Gen Z', 'Millennials', 'Gen X', 'Boomers']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
df['Age_Group_Code'] = df['Age_Group'].cat.codes
df.to_csv('../data/processed/featured_data.csv', index=False)
print("Feature Engineering complete. Data saved to 'featured_data.csv'.")
```
